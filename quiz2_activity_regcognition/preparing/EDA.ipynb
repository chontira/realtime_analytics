```python
from scipy.stats import zscore
import numpy as np
```

```python
z_scores = zscore(df[['x_sd',	'y_sd',	'z_sd']])
```

```python
z_scores
```

```python
abs_z_scores = np.abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
new_df = df[filtered_entries]
```
